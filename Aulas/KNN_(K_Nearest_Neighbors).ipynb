#Introdução ao KNN

As métricas de distância conseguem calcular a similaridade entre dois elementos.
Pode ser usado para: Classificação ou Regressão.

Distância Euclidiana

- Teorema de Pitagoras



```python
2-5
```

```python
(2-5)**2 #positivo
```

```python
((2-5)**2)**(0.5) #linha anterior elevada a 0.5 ou 1/2 ou raiz quadrada
```

```python
#Distância entre dois elementos com características diferentes a=[massa, cor]
a = [5,0.75]
b = [2,0.50]
```

Distância entre dois pontos separados por atributos [massa de a - massa de b] elevado ao quadrado e soma com a distância do próximo atributo [cor de a - cor de b] elevado ao quadrado. No final a raiz da soma das diferenças ao quadrado.

```python
((5-2)**2 + (0.75 - 0.50)**2)**0.5
```

#KNN para Classificação

Vamos importar o método KNN do sklearn.

```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
```

*n_neighbors=3*

Vizinhos próximos para os dados de treino, cada novo elemento será comparado aos 3 vizinhos mais próximos a ele.

Se a coluna de atributos tem valores numéricos o KNN não é aconselhado. Exemplo: o número 1 representa maçã e 2 banana.

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
data = pd.read_table('fruit_data_with_colors.txt')
```

```python
data
```

```python
X = data[['mass','height','width','color_score']] #atributos
y = data['fruit_label'] #rótulo
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

75% para treino e 25% para teste.

X - letra maiúscula para matriz.

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
knn.fit(X_train,y_train) #armazenamos os valores no knn
```

Os dados de treino e teste são aleatórios.

Bom acerto acima de 70 ou 80%.

A ordem de grandeza dos dados faz diferença na distância. As grandezas de cor e massa são muito diferentes.

```python
knn.score(X_test,y_test)
```

Para tentar balancear essa diferença podemo sutilizar a escala minmax.

#Pré-processando os dados para o KNN

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
```

Transformar numa escala de zero e um:

```python
X_train = mm.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = mm.transform(X_test) 
```

```python
X_test
```

```python
knn = KNeighborsClassifier(n_neighbors=3) #retreinar
```

```python
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test) #100% de acerto
```

```python
knn.predict(X_test)
```

```python
y_test
```
